```python
#  Projeto ETL - Bootcamp DIO
# Autor: Pablo Sena

# ============================================
# ETAPA 1 - IMPORTAÇÃO DE BIBLIOTECAS
# ============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# ETAPA 2 - EXTRAÇÃO DE DADOS
# ============================================
# Simulação de dados fictícios
dados = pd.DataFrame({
    "id_usuario": [101, 102, 103, 104, 105],
    "nome": ["Ana", "Bruno", "Carla", "Diego", "Eva"],
    "idade": [23, 35, 29, 42, 18],
    "cidade": ["Rio", "SP", "BH", "Curitiba", "Fortaleza"]
})

print("Dados extraídos:")
print(dados)

# ============================================
# ETAPA 3 - TRANSFORMAÇÃO DE DADOS
# ============================================
# Categorização por faixa etária
def classificar_idade(idade):
    if idade < 20:
        return "Adolescente"
    elif idade < 30:
        return "Jovem"
    elif idade < 40:
        return "Adulto"
    else:
        return "Sênior"

dados["faixa_etaria"] = dados["idade"].apply(classificar_idade)

# Normalização de nomes
dados["nome_normalizado"] = dados["nome"].str.upper()

print("\n Dados transformados:")
print(dados)

# ============================================
# ETAPA 4 - CARREGAMENTO DE DADOS
# ============================================
# Salvar como CSV no ambiente do Colab
dados.to_csv("dados_transformados.csv", index=False)

# Download do arquivo gerado
from google.colab import files
files.download("dados_transformados.csv")

# ============================================
# ETAPA 5 - VISUALIZAÇÃO DE DADOS
# ============================================
plt.figure(figsize=(8, 5))
sns.countplot(x="faixa_etaria", data=dados, palette="viridis")
plt.title("Distribuição por Faixa Etária")
plt.xlabel("Faixa Etária")
plt.ylabel("Quantidade de Usuários")
plt.show()
```
